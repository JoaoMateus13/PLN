```python
import nltk
from nltk.tokenize import word_tokenize
import string
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python

#Tokenização (word)
def tokenizar(caminho):
    
    with open(caminho, 'r', encoding='utf-8') as file:
        texto = file.read()

    palavras = word_tokenize(texto)

    return palavras

#Lower casing (conversão de todos os caracteres para minúsculo)
def lowerCase(palavras):
    
    result=[]

    for palavra in palavras:
        result.append(palavra.lower())

    return palavras

#Remoção de pontuações e outros caracteres especiais
def removerPontuacao(palavras):

    result = []

    for palavra in palavras:
        if palavra not in string.punctuation:
            result.append(palavra)

    return result



#Remoção de stopwords
def removerStopWord(palavras, stopwords):

    result = []

    for palavra in palavras:
        if palavra not in stopwords:
            result.append(palavra)

    return result


#Remoção das n palavras mais frequentes
def removerPMaisFrequentes(palavras, n):

    result = []
    
    freq = FreqDist(palavras)

    palavras_frequentes = freq.most_common(n)

    palavras_frequentes = [palavra for palavra, _ in palavras_frequentes]

    for palavra in palavras:
        if palavra not in palavras_frequentes:
            result.append(palavra)

    
    return result

#Remoção das palavras que ocorreram mais que n vezes
def removerNMaisFrequentes(palavras, n):
        
    result = []

    freq = FreqDist(palavras)


    for palavra, frequen in freq.items():
        if frequen > n:
            result.append(palavra)

    return result


#Remoção das n palavras menos frequentes
def removerPMenosFrequentes(palavras, n):

    result = []

    freq = FreqDist(palavras)

    palavras_frequentes = freq.most_common()[-n:]

    palavras_frequentes = [palavra for palavra, _ in palavras_frequentes]

    for palavra in palavras:
        if palavra not in palavras_frequentes:
            result.append(palavra)

    return result

#Remoção das palavras q ue ocorreram menos que n vezes
def removerNMenosFrequentes(palavras, n):
    result = []

    freq = FreqDist(palavras)


    for palavra, frequen in freq.items():
        if frequen < n:
            result.append(palavra)

    return result


#Quantidade de types
def contador_Types(caminho):

    with open(caminho, 'r', encoding='utf-8') as file:
        texto = file.read()

    palavras = word_tokenize(texto)

    result = []

    for palavra in palavras:
        if palavra not in string.punctuation:
            result.append(palavra)

    
    return len(set(result))

#quantidade de tokens
def contador_Tokens(caminho):

    with open(caminho, 'r', encoding='utf-8') as file:
        texto = file.read()

    palavras = word_tokenize(texto)

    result = []

    for palavra in palavras:
        if palavra not in string.punctuation:
            result.append(palavra)

    
    return len(result)


#Ranking dos 20 types mais frequentes no texto
def ranking_20(palavras):
    
    freq = FreqDist(palavras)

    plt.figure(figsize=(8, 4))
    freq.plot(20, title="Frequência de Palavras")
```



## Função <small>`tokenizar(caminho)`</small>
Esta função lê um arquivo especificado pelo parâmetro `caminho` e tokeniza seu conteúdo usando a função `word_tokenize` da biblioteca NLTK. Ela retorna uma lista de palavras (tokens) no texto.

## Função <small>`lowerCase(palavras)`</small>
Esta função recebe uma lista de palavras (`palavras`) como entrada e converte todos os caracteres de cada palavra para minúsculas. Ela retorna a lista de palavras em que todas as letras estão em minúsculas.

## Função <small>`removerPontuacao(palavras)`</small>
Dada uma lista de palavras (`palavras`), esta função remove pontuações e outros caracteres especiais de cada palavra. Ela retorna uma lista de palavras sem os caracteres de pontuação.

## Função <small>`removerStopWord(palavras, stopwords)`</small>
Esta função recebe uma lista de palavras (`palavras`) e uma lista de stopwords (`stopwords`) como entrada. Ela remove as palavras que estão na lista de stopwords da lista de palavras. A função retorna uma lista de palavras sem as stopwords.

## Função <small>`removerPMaisFrequentes(palavras, n)`</small>
Ela remove as n palavras mais frequentes da lista de palavras fornecida (`palavras`) com base em sua frequência de ocorrência. A função utiliza a biblioteca NLTK para calcular a frequência das palavras. O resultado é uma lista de palavras sem as n palavras mais frequentes.

## Função <small>`removerNMaisFrequentes(palavras, n)`</small>
Similar à função anterior, esta função remove as palavras que ocorrem mais do que n vezes na lista de palavras fornecida. A frequência de ocorrência é calculada usando a biblioteca NLTK. A função retorna uma lista de palavras sem as palavras que ocorrem mais de n vezes.

## Função <small>`removerPMenosFrequentes(palavras, n)`</small>
Esta função remove as n palavras menos frequentes da lista de palavras com base em sua frequência de ocorrência. A frequência é calculada usando a biblioteca NLTK. O resultado é uma lista de palavras sem as n palavras menos frequentes.

## Função <small>`removerNMenosFrequentes(palavras, n)`</small>
Semelhante à função anterior, esta função remove as palavras que ocorrem menos do que n vezes na lista de palavras fornecida. A frequência de ocorrência é calculada usando a biblioteca NLTK. O resultado é uma lista de palavras sem as palavras que ocorrem menos de n vezes.

## Função <small>`contador_Types(caminho)`</small>
Esta função lê um arquivo especificado pelo parâmetro `caminho`, tokeniza o texto e conta o número de tipos (palavras únicas) no texto.

## Função <small>`contador_Tokens(caminho)`</small>
Esta função lê um arquivo especificado pelo parâmetro `caminho`, tokeniza o texto e conta o número total de tokens (palavras) no texto.

## Função <small>`ranking_20(palavras)`</small>
Esta função cria um gráfico de barras mostrando o ranking das 20 palavras mais frequentes na lista de palavras fornecida (`palavras`) usando a biblioteca NLTK (FreqDist) e Matplotlib para visualização.


```python
caminho = "database\capivara-pt.txt"
stopwords= tokenizar("database\stopwords-pt.txt")
```

```python
print(tokenizar(caminho))
print(lowerCase(tokenizar(caminho)))
print(removerPontuacao(lowerCase(tokenizar(caminho))))
print(removerStopWord(removerPontuacao(lowerCase(tokenizar(caminho))),stopwords))


removerNFrequentes = removerStopWord(removerPontuacao(lowerCase(tokenizar(caminho))),stopwords)

print(removerPMaisFrequentes(removerNFrequentes,20))
print(removerPMenosFrequentes(removerNFrequentes,20))

print(removerNMaisFrequentes(removerNFrequentes,1))
print(removerNMenosFrequentes(removerNFrequentes,4))

print(contador_Tokens(caminho))
print(contador_Types(caminho))

print(ranking_20(removerNFrequentes))


```
